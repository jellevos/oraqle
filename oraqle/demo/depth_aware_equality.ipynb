# A new paradigm in arithmetization: Depth-aware arithmetization

#### An equality circuit is easy to define but somewhat hard to optimize!

```python
from galois import GF

from circuit_compiler.compiler.nodes.leafs import Input
from circuit_compiler.compiler.comparison.equality import Equals
from circuit_compiler.compiler.circuit import Circuit

gf = GF(467)

a = Input("a", gf)
b = Input("b", gf)

output = Equals(a, b, gf)

circuit = Circuit(outputs=[output], gf=gf)
circuit.display_graph()
```

#### Naive methods only find only one arithmetization for this high-level circuit.

```python
naive_arithmetic_circuit = circuit.arithmetize("naive")
naive_arithmetic_circuit.display_graph()
```

#### Our depth-aware arithmetization method finds a second circuit that is potentially faster!

```python
depth_aware_circuits = circuit.arithmetize_depth_aware(0.5)

for depth, cost, depth_aware_circuit in depth_aware_circuits:
    print("Depth: ", depth, ", ", "Cost: ", cost)
    depth_aware_circuit.display_graph()
```
